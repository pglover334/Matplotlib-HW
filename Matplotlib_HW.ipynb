```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import pandas as pd
import numpy as np
```

```python
city_data = "C:\\Users\\User\\city_data.csv"
ride_data = "C:\\Users\\User\\ride_data.csv"
```

```python
df1 = pd.read_csv(city_data)
df1.head()
```

```python
df2 = pd.read_csv(ride_data)
df2.head()
```

```python
combined_df = pd.merge(df1, df2, how='outer', on='city')
combined_df.head()
```

```python
combined_df.groupby("type").head()
```

```python
y_axis = combined_df.groupby(['city'], as_index=False).fare.mean()
y_axis
```

```python
s_value = combined_df.groupby(['city'], as_index=False).driver_count.count()
s_value
```

```python
x_axis = combined_df.groupby(['city'], as_index=False).ride_id.count()
x_axis
```

```python
y_axis2 = y_axis['fare'].tolist()
y_axis2
```

```python
s_value2 = s_value['driver_count'].tolist()
s_value2
```

```python
x_axis2 = x_axis['ride_id'].tolist()
x_axis2
```

```python
plt.scatter(x_axis2, y_axis2, marker="o", color="blue", facecolors="red", edgecolors="black", s=s_value2)
plt.grid()
plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.show()
```

```python
colorstype=df1[['city','type']]
```

```python
x_axis44 = combined_df.groupby(['city'], as_index=False).ride_id.count()
x_axis44
```

```python

plotcolors=pd.merge(x_axis44, colorstype, how='left', on='city')
plotcolors
```

```python
colors2=plotcolors['type'].tolist()
colors2
```

```python
finalcolors=[]
for a in colors2:
    if a=='Urban':
        finalcolors.append('coral')
    elif a=='Suburban':
        finalcolors.append('skyblue')
    elif a=='Rural':
        finalcolors.append('gold')
        
finalcolors      
```

```python
#Bubble Plot for submission

plt.scatter(x_axis2, y_axis2, s=s_value2, c=finalcolors, alpha=1, edgecolor="black", linewidth=.2)
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.figtext(1.22,.6,'Note: Circle size correlates with driver count per city.',fontsize=10,ha='center')

legend_1 = mlines.Line2D([], [], color='coral', marker='o',
                          markersize=8, label='Urban')
legend_2 = mlines.Line2D([], [], color='skyblue', marker='o',
                          markersize=8, label='Suburban')
legend_3 = mlines.Line2D([], [], color='gold', marker='o',
                          markersize=8, label='Rural')
plt.legend(title="City Types", handles=[legend_1, legend_2, legend_3])

plt.savefig("C:\\Users\\User\\Bubble_Plot_Matplotlib_HW.png")
plt.grid()
plt.show()
```
